## Present Value Formula

The present value of a future cash flow is:

$$PV = \frac{FV}{(1 + r)^n}$$

Where:
- $FV$ = future value
- $r$ = interest rate (per period)
- $n$ = number of periods

```python
# Example: What is the present value of $1,000 received in 5 years at 4% per year?

fv = 1000
r = 0.04
n = 5

pv = fv / (1 + r)**n
print(f"Present value: ${pv:.2f}")
```

The output shows that $1,000 in 5 years at 4% is worth about **$821.93** today.

If you change `r` or `n`, you can see how the present value changes.

```python
# Try changing these values and re-running the cell

fv = float(input("Future value ($): ") or 1000)
r = float(input("Annual interest rate (e.g., 0.04 for 4%): ") or 0.04)
n = int(input("Number of years: ") or 5)

pv = fv / (1 + r)**n
print(f"Present value: ${pv:.2f}")
```

## Present Value of Multiple Cash Flows

If you receive cash flows over several years, you sum the PV of each:

$$PV = \sum_{t=1}^{n} \frac{CF_t}{(1 + r)^t}$$

Where $CF_t$ is the cash flow at time $t$.

```python
# Example: $100 per year for 3 years at 5%

cash_flows = [100, 100, 100]
r = 0.05

pv_total = 0
for t, cf in enumerate(cash_flows, start=1):
    pv = cf / (1 + r)**t
    pv_total += pv
    print(f"Year {t}: PV = ${pv:.2f}")

print(f"Total PV: ${pv_total:.2f}")
```
